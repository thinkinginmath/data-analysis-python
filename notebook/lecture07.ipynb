```python
class LetterCounter():
    '''Counts letters in a text stream'''
    def __init__(self, letter):
        self.letter=letter
        self.count=0
    def increment(self):
        self.count+=1
    def process_file(self, filename):
        with open(filename, 'r') as f:
            for line in f:
                self.process_line(line)
    def process_line(self, line):
        for x in line:
            if x==self.letter:
                self.increment()
    def get_count(self):
        return self.count
lc = LetterCounter('e')
fname = '/Users/keith/Downloads/mobydick.txt'
lc.process_file(fname)
lc.get_count()
```

```python
def count_letter(fname, letter):
    with open(fname, 'r') as f:
        return(sum([c==letter
                    for line in f
                    for w in line
                    for c in w]))
count_letter(fname, 'e')
```

```python
def quicksort(t):
    if len(t) <= 1:
        return t # list is already sorted.
    else:
        pivot = t[0]
        return(quicksort([x for x in t if x < pivot]) +
               [x for x in t if x==pivot] +
              quicksort([x for x in t if x > pivot]))
quicksort([3,5,4,2,1,6,5,7,4,0,0,2,4])
```

```python
def tuple_add(x,y):
    if len(x) != len(y):
        raise TypeError('Tuple lengths must agree')
    r = len(x)*[0]
    for i in range(len(x)):
        r[i] = x[i] + y[i]
    return tuple(r)
functools.reduce(tuple_add, [(1,2),(1,3),(2,5),(3,7),(5,11)])
```

```python
def compose(*funcs):
    '''Return a new function that is the
    composition of the argument functions.'''
    def inner(data, funcs=funcs):
        result = data
        for f in reversed(funcs):
            result = f(result)
        return result
    return inner
f = lambda x: x**2
g = lambda x: x+1
# compose(g,f) == g(f(x)) == x**2 + 1
list(map(compose(g,f), range(10)))
```
